```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.metrics import r2_score, accuracy_score
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

## XGBoost Classifier

```python
# Load in the wine dataset
wine = datasets.load_wine()
```

```python
# Create the wine `data` dataset as a dataframe and name the columns with `feature_names`
df = pd.DataFrame(data=wine['data'], columns=wine['feature_names'])

# Include the target as well
df['target'] = wine['target']
```

```python
# Split your data with these ratios: train: 0.8 | test: 0.2
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0)
```

```python
# Load your train/test dataframe into DMatrix
dtrain = xgb.DMatrix(df_train.drop("target", axis=1), label=df_train['target'])
dtest = xgb.DMatrix(df_test.drop("target", axis=1), label=df_test['target'])
```

```python
# How does the model perform on the training dataset and default model parameters?
# Using the hyperparameters in the requirements, is there improvement?
# Remember we use the test dataset to score the model
params = {
    'max_depth': 5,
    'eta': 0.1,
    'objective': 'multi:softmax',
    'num_class': 3
}

num_round = 100
bst = xgb.train(params, dtrain, num_round)
```

The `num_round` parameter in XGBoost refers to the number of boosting rounds or iterations. It specifies the number of trees to be built in the XGBoost model. Each boosting round adds a new tree to the model, and the model learns from the mistakes made in the previous rounds.

The `num_round` parameter is set to 100, which means that the XGBoost model will be trained using 100 boosting rounds. This means that 100 decision trees will be created and added to the model to iteratively improve its performance.

Adjusting the `num_round` parameter can have an impact on the model's performance. Increasing the number of rounds can potentially improve the model's ability to capture complex patterns in the data but may also lead to overfitting if not properly regularized. On the other hand, reducing the number of rounds can speed up training but may result in underfitting and lower model performance.

```python
# xgboost is not scikit learn, so you'll need to do predictions using their API
preds = bst.predict(dtest)
```

```python
# Accuracy score using scikit learn function for classification metric
accuracy_score(y_true=dtest.get_label(), y_pred=preds)
```

```python
# Plot the importance of the features based on fitted trees
xgb.plot_importance(bst)
```

## XGBoost Regression

```python
# Load in the diabetes dataset
diabetes = datasets.load_diabetes()
```

```python
# Create the diabetes `data` dataset as a dataframe and name the columns with `feature_names`
dfd = pd.DataFrame(data=diabetes['data'], columns=diabetes['feature_names'])

# Include the target as well
dfd['target'] = diabetes['target']
```

```python
# Split your data with these ratios: train: 0.8 | test: 0.2
dfd_train, dfd_test = train_test_split(dfd, test_size=0.2, random_state=0)
```

```python
# Load your train/test dataframe into DMatrix
dtrain = xgb.DMatrix(dfd_train.drop('target', axis=1), label=dfd_train['target'])
dtest = xgb.DMatrix(dfd_test.drop('target', axis=1), label=dfd_test['target'])
```

```python
# How does the model perform on the training dataset and default model parameters?
# Using the hyperparameters in the requirements, is there improvement?
# Remember we use the test dataset to score the model

params = {
    'max_depth': 2,
    'eta': 0.03,
    'gamma': 0.09,
    'colsample_bytree': 0.5,
    'objective': 'reg:squarederror'
}

num_round = 100
bst = xgb.train(params, dtrain, num_round)
```

```python
# xgboost is not scikit learn, so you'll need to do predictions using their API
preds = bst.predict(dtest)
```

```python
# R2 score using scikit learn function for regression metric
r2_score(y_true=dtest.get_label(), y_pred=preds)
```

```python
# Plot the importance of the features based on fitted trees
xgb.plot_importance(bst)
```
